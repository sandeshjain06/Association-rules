 # Association Rules - Data Mining (Books)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
books=pd.read_csv("D:\\Sandesh\\Data Science\\Homework Assignment\\Machine Learning\\Association Rules\\book.csv")
books.head()
```

 * 1 ) Calculate Association Rules (1% support, 95% confidence)


```python
frequent_items=apriori(books,min_support=0.1,use_colnames=True)
frequent_items
```

```python
assoc_rules=association_rules(frequent_items,metric='lift',min_threshold=0.95)
assoc_rules
```

```python
assoc_rules[assoc_rules.lift>1]
```

```python
plt.scatter(assoc_rules['support'],assoc_rules['confidence'])
plt.xlabel('Support')
plt.ylabel("Confidence")
plt.show()
```

#   Association Rules (5% support, 90% confidence)

 # Association Rules - Data Mining (Movies)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
movies=pd.read_csv("D:\\Sandesh\\Data Science\\Homework Assignment\\Machine Learning\\Association Rules\\my_movies.csv")
movies.head()
```

```python
movies_new=movies.iloc[:,5:]
movies_new
```

 # Support 5% , Confidence 90%

```python
moviesfreqitems=apriori(movies_new,min_support=0.05,use_colnames=True)
moviesfreqitems.head()
```

```python
model=association_rules(moviesfreqitems,metric='lift',min_threshold=0.90)
model.head()
```

```python
model[model.lift>1]
```

```python
plt.scatter(model['support'],model['confidence'])
plt.show()
```

#  Support 10% and confidence 70%


```python
freq_items2=apriori(movies_new,min_support=0.10,use_colnames=True)
freq_items2.head()
```

```python
model2=association_rules(freq_items2,min_threshold=0.70,metric='lift')
model2.head()
```

```python
model2[model2.lift>1]
```

```python
plt.scatter(model2['support'],model2['confidence'])
plt.show()
```
